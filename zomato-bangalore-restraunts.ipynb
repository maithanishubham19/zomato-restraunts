```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df=pd.read_csv('/kaggle/input/zomato-bangalore-restaurants/zomato.csv')
df.head()
```

```python
df.columns
df.shape
```

```python
df=df.drop(['url','address','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
df.info()
df.shape
```

```python
mask=df.duplicated()
mask
```

```python
df[mask]
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

```python
def handle_rate(value):
    if(value=='NEW' or value =='-'):
        return np.nan
    else:
        value=str(value).split("/")  #split will store the rating in an array
        value=value[0]     # value[0] will have the 4.1 part where as value[1]=5
        return float(value)

df['rate']=df['rate'].apply(handle_rate)
df['rate'].head()
        
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isna().sum() #check the no of null values
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.head()
```

**rename long columns**

```python
df.rename(columns={'approx_cost(for two people)':'cost2plate','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df = df.drop(['listed_in(city)'],axis=1)
df
```

```python
df['cost2plate'].unique()
```

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',',"")
        return float(value)
    else:
        return float(value)

df['cost2plate']=df['cost2plate'].apply(handlecomma)
df['cost2plate'].unique()
    
```

```python
df['rest_type'].value_counts()
```

```python
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types  #stores the rest_type
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

make a fn to club restypes less than 1000 as others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
    
df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts()
location
```

```python
location_lessthan300=location[location<300]
location_lessthan300
```

```python
def handle_location(value):
    if value in location_lessthan300:
        return 'others'
    else:
        return(value)
    
df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
df['location'].unique()
```

```python
df['cuisines'].value_counts()
```

```python
cuisines=df['cuisines'].value_counts()
cuisines_lessthan100=cuisines[cuisines<300]

def handle_cuisine(value):
    if value in cuisines_lessthan100:
        return 'others'
    else:
        return value
    
df['cuisines']=df['cuisines'].apply(handle_cuisine)
df['cuisines'].value_counts()
```
